```python
df['Order Date']=pd.to_datetime(df['Order Date'])
df['Ship Date']=pd.to_datetime(df["Ship Date"])
```

```python
df.head(2)
```

```python
df['Order Month']=df['Order Date'].dt.month
df['Order Year']=df['Order Date'].dt.year
df['Order Day of week'] = df['Order Date'].dt.dayofweek
```

```python
df.head()
```

```python
sales = df.groupby('Order Month')['Sales'].sum().reset_index()
fig = px.line(sales,x='Order Month',y = "Sales",title="Monthly Sales")
fig.show()
```

```python
sales = df.groupby('Order Date')['Sales'].sum().reset_index()
sales
```

```python
sales_by_category = df.groupby("Category")["Sales"].sum().reset_index()
sales_by_category
```

```python
sales_by_category = df.groupby("Category")["Sales"].sum().reset_index()
fig = px.pie(sales_by_category,values="Sales",names="Category",hole=0.5,title="Sales by Category",color_discrete_sequence=px.colors.qualitative.Pastel) # Changed pastel to Pastel
fig.update_traces(textposition = "inside",textinfo="percent+label")
fig.update_layout(title_font=dict(size=24))
fig.show()
```

```python
sales_by_sub_category = df.groupby("Sub-Category")["Sales"].sum().reset_index()
fig = px.bar(sales_by_sub_category,x="Sub-Category",y="Sales",title="Sales by Sub-Category")

fig.show()
```

```python
df.head()
```

```python
monthly_profit = df.groupby("Order Month")["Profit"].sum().reset_index()
fig = px.line(monthly_profit,x="Order Month",y ="Profit",title="Monthly Profit")
fig.show()
```

```python
monthly_profit = df.groupby("Order Month")["Profit"].sum().reset_index()
fig = px.bar(monthly_profit,x="Order Month",y ="Profit",title="Monthly Profit")
fig.show()
```

```python
profit_by_category = df.groupby("Category")["Profit"].sum().reset_index()
fig = px.pie(profit_by_category,values="Profit",names="Category",hole=0.5,title="Profit by Category")
fig.update_traces(textposition ="inside",textinfo ="percent+label")
fig.update_layout(title_font = dict(size=24))
fig.show()
```

```python
profit_by_sub_category = df.groupby("Sub-Category")["Profit"].sum().reset_index()
fig = px.bar(profit_by_sub_category,x="Sub-Category",y="Profit",title="Profit by Sub-Category")
fig.show()
```

```python
df.head()
```

```python
sales_profit_by_segment = df.groupby("Segment").agg({"Sales":"sum","Profit":"sum"}).reset_index()
color_palette = px.colors.qualitative.Pastel
fig= go.Figure()
fig.add_trace(go.Bar(x=sales_profit_by_segment["Segment"],y=sales_profit_by_segment["Sales"],name="Sales",marker_color=color_palette[0]))
fig.add_trace(go.Bar(x=sales_profit_by_segment["Segment"],y=sales_profit_by_segment["Profit"],name="profit",marker_color  = color_palette[1]))
fig.update_layout(title="Sales and Profit by Segment",xaxis_title="Segment",yaxis_title="Amount")
fig.show()
```
