```python
#Data Cleaning
df.isnull().sum()
#dropping all missing values rows
df.dropna(inplace=True)
df
```

```python
# Checking if there is any duplicated
df.duplicated().sum()
#Dropping the duplicated
df.drop_duplicates(inplace=True)
```

```python
df.duplicated().sum()
```

```python
# changing data types
df.dtypes
df['id']=df['id'].astype(object)
df['host_id']=df['host_id'].astype(object)
```

```python
#identifying is there any outliers and fix for now
df1=df[df['price'] < 1500]
plt.figure(figsize=(8,5))
sns.boxplot(data=df1,x='price')
```

```python
# price dependency on negibourhood
plt.figure(figsize=(8,5))
sns.barplot(data=df1,x='neighbourhood_group',y='price',hue='room_type')
plt.title("Price dependency on negibourhood")
plt.show()
```

```python
#number of reviews and price
plt.figure(figsize=(8,5))
sns.scatterplot(data=df1,x='number_of_reviews',y='price',hue='neighbourhood_group')
plt.title("number of reviews and price")
plt.show()
```

```python
sns.pairplot(data=df1,vars=['price','minimum_nights','number_of_reviews','availability_365'],hue='room_type')
```
